```python
belgium = visitors.query('market=="Belgium" or market=="Saudi Arabia" or market=="USA"')
```

```python
belgium = belgium.groupby(['market','year']).mean()
```

```python
belgium = belgium.reset_index()
```

```python
belgium['visits_spend'] = belgium['spend']/belgium['visits']
```
